```python
#IMPORTING REQUIRED MODULES

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
#DATA IS IMPORTED

data=pd.read_csv('real_estate_price_size.csv')
```

```python
y=data["price"] #dependentvariable
x1=data["size"] #independentvariable
```

```python
#EXPLORE THE DATA AND CREATING REGRESSION

plt.scatter(x1,y)
yhat=223.1787*x1+101900 #yhat is determined with the obtained coefficients by equating size in (x1)
fig=plt.plot(x1,yhat,lw=4,c='orange',label="regression line")
plt.xlabel("size",fontsize=15,color="red")
plt.ylabel("price",fontsize=15,color="blue")
plt.show()
```

```python
#REGRESSION ITSELF

x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```
